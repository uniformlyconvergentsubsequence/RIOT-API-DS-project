```python
#Creates a dictionary which contains the roles of all champions
champ_role_dict = {}
champ_df = pd.read_excel("Champs_w_roles.xlsx")
role_list = ['Top', 'Jungle', 'Mid', 'Bot', 'Sup']
for num in range(len(champ_df.get('name'))):
    champ_name = champ_df.get('name')[num]
    #cycles through all roles and adds to dictionary if the champ plays that role
    for role in role_list:
        if type(champ_df.get(role)[num]) != float:
            if champ_name not in champ_role_dict:
                champ_role_dict[champ_name] = [role]
            else:
                champ_role_dict[champ_name].append(role)
champ_role_dict
```

```python
#Iterates through a dictionary (e.g account_list) to provide a dictionary total for all accounts in the list
def pro_dict_to_mastery(pros, role):
    percentage_dict = {}
    mastery_dict = {}
    mastery_dict['total'] = 0
    #iterates through reigon
    for region in list(pros.keys()): 
        rr = region.replace(':', '')
        #iterate through account name and makes a 
        for account in list(pros[region]):
            ss = ra.get_mastery_by_summoner_name(account, rr)
            for num in range(len(ss.get('championName'))):
                champ = ss.get('championName')[num+1]
                champ_points = ss.get('championPoints')[num+1]
                # If the champion has not been played in another account, add it directly to the dictionary
                # If it has been played, update the dictionary's points by adding the points for this acc
                if champ not in list(mastery_dict.keys()): 
                    mastery_dict[champ] = champ_points
                else: 
                    mastery_dict[champ] += champ_points
    mastery_dict['total'] = sum(list(mastery_dict.values()))
    onetrickscore = 0
    mainrolescore = 0
    percentage_dict['onetrickscore'] = onetrickscore
    # creates a new dictionary where each champion key corresponds to a list 
    # featuring mastery points as 1st value and % total mastery points as 2nd value
    for champ in list(mastery_dict.keys()):
        
        #Onetrickscore is calculated by mean square error of the percentage_mastery of that champ 
        # vs the "expected" average of playing all 165 champs equally
        if champ != 'total':
            onetrickscore += (100 * mastery_dict[champ]/mastery_dict['total'] - (100/165))**2
        #if the champ is played in the player's main role, add the mastery points to the mainrolescore, then divide at the end
            if role in champ_role_dict[champ]:
                mainrolescore += mastery_dict[champ]
    mainrolepercent = mainrolescore/mastery_dict['total']
    percentage_dict['onetrickscore'] = onetrickscore
    #returns a list featuring mainrolescore, onetrickscore, and total mastery points
    return [mainrolepercent, onetrickscore, mastery_dict['total']]
```

```python
#This goes through the entire pro dataframe and runs the function on each pro player and their role
# It then adds the mainrolescore, onetrickscore, and totalmastery to the dataframe
# This takes a while to run
# It also saves the dataframe to a csv file called pro_mastery.csv
# This is the final dataframe that we will use for the project
# 
list_of_all_pros = []
mainrolescore = []
onetrickscore = []
totalmastery = []
for num in range(len(df.get('Account_list'))) : 
    try: list_of_all_pros.append(pro_dict_to_mastery(df.get('Account_list')[num], df.get('Roles (LeaguePedia)')[num]))
    except: list_of_all_pros.append([0, 0, 0])
for pro in list_of_all_pros:
    mainrolescore.append(pro[0])
    onetrickscore.append(pro[1])
    totalmastery.append(pro[2])
df['mainrolescore'] = mainrolescore
df['onetrickscore'] = onetrickscore
df['totalmastery'] = totalmastery
df.to_csv("pro_mastery.csv")
df
```
